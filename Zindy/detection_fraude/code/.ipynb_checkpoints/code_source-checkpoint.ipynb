```python
#sépartion des variables (X) et de la cible (y)
X = df_train.drop(['target', 'id'], axis=1)
y = df_train['target']
```

```python
#séparation Apprentissage/Validation
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size = 0.8, random_state = 0)
```

```python
num_columns = X.select_dtypes(exclude=['object'])
cat_columns = X.select_dtypes(include=['object'])
```

```python
num_columns = list(num_columns)
cat_columns = list(cat_columns)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer, make_column_selector as selector
import numpy as np
 
numeric_tranformer = Pipeline(steps =[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
]
)

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy= 'most_frequent')),
    ('hot', OneHotEncoder(handle_unknown= 'ignore'))
])

preprocessor = ColumnTransformer(transformers =[
    ('num', numeric_tranformer, num_columns),
    ('cat', categorical_transformer, cat_columns)
])
```

```python
#definition du modèle 
from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor

linear_regression = LinearRegression()
sgd_regression = SGDRegressor()
tree = DecisionTreeRegressor()
random_forest = RandomForestRegressor()
xgboost = XGBRegressor()

modele = {'linear_regression' : linear_regression , 'sgd_regression': sgd_regression, 'tree' : tree, 
          'random_forest' : random_forest, 'xgboost' : xgboost}
```

```python
modele.items()
```

```python
from sklearn.metrics import mean_squared_error

for name, model in modele.items():
    my_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('modele', model)
])     
    my_pipeline.fit(X_train, y_train)
    y_predict = my_pipeline.predict(X_valid)
    print(name, 'mean_square_error: ', mean_squared_error(y_predict, y_valid), 'score :', my_pipeline.score(X_valid, y_valid))
```

```python
my_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('modele', RandomForestRegressor())
])

my_pipeline.fit(X_train, y_train)
```

```python
my_pipeline.get_params()
```

```python
#random forest est le meilleur modèle pour ce jeu de données.
#optimisation du modèle radom forest
from sklearn.model_selection import GridSearchCV

my_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('modele', RandomForestRegressor())
])  

grid_param = {'modele__n_estimators': [100, 400, 600, 100], 'preprocessor__num__imputer__strategy': ['mean','median']}

grid = GridSearchCV(my_pipeline, grid_param, cv = 5)

grid.fit(X_train, y_train)
```

```python
#meilleurs parametres
grid.best_params_
```

```python
#meilleur score
grid.best_score_
```

```python
best_model = grid.best_estimator_
df_test = df_test.drop('id', axis=1)
prediction_test = best_model.predict(df_test)
```
